```python
import pandas as pd
import random


from feature_extraction import (
    checkIP, urlLength, url_shortning, check_At, check_double_slash,
    prefix_suffix, httpDomain, getDepth, no_dns_record,
    whois_registered, find_domain_age, page_rank
)


input_file = "/content/drive/MyDrive/AML/Final project/phishing_phishtank.csv"

df = pd.read_csv(input_file)
urls = df['url']
# sampled_urls = urls.sample(n=min(50, len(urls)), random_state=42)
sampled_urls = urls.head(7000)


results = []
count = 0


for url in sampled_urls:
    if count % 100 == 0:
      print("----------------------------------------------",count)
    count += 1
    try:
        result = {
            "Input_URL": url,
            "Have_IP": checkIP(url),
            "URL_Length": urlLength(url),
            "TinyURL": url_shortning(url),
            "Have_At": check_At(url),
            "Redirection": check_double_slash(url),
            "Prefix/Suffix": prefix_suffix(url),
            "https_Domain": httpDomain(url),
            "URL_Depth": getDepth(url),
            "DNS_Record": no_dns_record(url),
            "Whois_Registered": whois_registered(url),
            "Domain_Age": find_domain_age(url),
            "Page_Rank": page_rank(url),
            "status": 1,
        }
        results.append(result)
    except Exception as e:
        print(f"Error processing URL {url}: {e}")


output_df = pd.DataFrame(results)


```

```python
# Save results to Excel
output_df.to_csv("url_features_output.csv", index=False)
```
